```python
import igraph as ig
import json
from bipdyngraph.DotLayoutParser import DotLayoutParser
from bipdyngraph.BipDynGraph import BipartiteDynGraph
from bipdyngraph.PyDotBipartiteLayout import PydotBipartiteLayout
```

```python
fp = "../data/Dufournaud_geoloc_1940.json"
with open(fp) as file:
    json_data = json.load(file)

graph = BipartiteDynGraph(json_data)

layout_bipartite = PydotBipartiteLayout(graph, same_person_edge_weight=10)
layout_bipartite.run()
```

```python
parser = DotLayoutParser(layout_bipartite, graph)
```

```python
parser.build_nx_graph()
layout_graph = parser.graph
```

```python
layout_graph.nodes.data()
```

```python
one_rank =layout_graph.nodes.data("rank")["m1"]
type(one_rank)
```

```python
igraph = ig.Graph.from_networkx(layout_graph)
```

```python
igraph.vertex_attributes()
```
